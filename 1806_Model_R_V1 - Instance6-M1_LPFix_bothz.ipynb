## Mathematical Model Formulation

```python
#Define Model
m = Model('RRB')
```

```python
## Time constraint of 3 hours
```

```python
m.setParam('TimeLimit',3*60*60)
```

```python
## Define x(i,j,k) as binary variable
```

```python
x={}
#y={}
for i in range(0,n):
    for j in range(0,n):
        if i != j:
            #y[i,j]=m.addVar(vtype=GRB.BINARY, name='y')
            for k in commodity_len:
                x[i,j,k]=m.addVar(vtype=GRB.BINARY, name='x')
```

```python
## Define y(i,j) as binary variable
```

```python
y = m.addVars(n, n, vtype=GRB.BINARY, name='y')
```

```python
## Define z(i,j,t) as binary variable
```

```python
z={}
for i in range(0,n):
    for t in range(0,len(track_i[i])):
        for j in range(0,n):
            if i != j:
                z[i,j,t]=m.addVar(vtype=GRB.BINARY, name='z')
            
            #for t in range(0,len(track_i[i])):

                    
```

## Objective Function

```python
m.setObjective(
    quicksum(quicksum(quicksum(pairs[i,j] * x[i,j,k] * car_numbers[com_rev[k]] * C for k in com_rev) for i in range(0,n) if i != j) for j in range(0,n)) + 
    quicksum(quicksum(quicksum(x[i,j,k] * 50 * C  for k in com_rev) for i in range(0,n) if i != j) for j in range(0,n))
) 
```

## Constraints

### 1. Flow conservation in blocking network

```python
for k in commodity_len:
    for i in range(0,n):
        if i == origin[com_rev[k]]:
            m.addConstr(quicksum(x[i,j,k] for j in range(0,n) if j!= i) - (quicksum(x[j,i,k] for j in range(0,n) if j!= i)) == 1)
        elif i == dest[com_rev[k]]:
            m.addConstr(quicksum(x[i,j,k] for j in range(0,n) if j!= i) - (quicksum(x[j,i,k] for j in range(0,n) if j!= i)) == -1)
        elif i != origin[com_rev[k]] or i != dest[com_rev[k]]:
            m.addConstr(quicksum(x[i,j,k] for j in range(0,n) if j!= i) - (quicksum(x[j,i,k] for j in range(0,n) if j!= i)) == 0)
            
        
```

### 2. If a block is not built for the path from  to  then the number of commodities assigned is 0. If the block is built, then the total number of commodities in the block can be a value in the range of 0 and the total number of commodities

```python
for i in range(0,n):
    for j in range(0,n):
        if j!=i:
            m.addConstr(quicksum(x[i,j,k] for k in commodity_len) <= len(commodity_len)*y[i,j])
```

### 3. Constraints on the maximum number of reclassification tracks available in node 

```python
for i in range(0,n):
    m.addConstr(quicksum(y[i,j] for j in range(0,n) if j != i) <= b[i])
```

### 4. Constraints considering that one block can span multiple tracks and guaranteeing that the total length of the commodities reclassified at node  does not exceed the length of the total tracks used

```python
for i in range(0,n):
    #if avg_len[i] != 0:
    for j in range(0,n):
        if j !=i:
            m.addConstr(quicksum(x[i,j,k]*l_k[k] for k in commodity_len) <= avg_len[i]*quicksum(z[i,j,t] for t in range(0,len(track_i[i]))))
       
        
```

```python
avg_len
```

### 5. Constraints ensuring that each track is used only by one block

```python
for i in range(0,n):
    for t in range(0,len(track_i[i])):
        m.addConstr(quicksum(z[i,j,t] for j in range(0,n) if j!=i) <= 1)         
    
```

### 6. Constraints (step 2A): There should be at least h(=5) cars in each block

```python
for i in range(0,n):
    for j in range(0,n):
        if i!=j:
            m.addConstr(quicksum(x[i,j,k] * car_numbers[com_rev[k]] for k in commodity_len) >= 5 * y[i,j]) 
```
